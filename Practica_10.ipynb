NOTA: Los datos del diccionario son los Datos básicos de un trabajador de una fabrica.


```python
def agregar_automatico(lista):
    pass

def agregar_manual(lista):
    id_trabajador = input("Ingrese el ID del trabajador: ")
    nombre = input("Ingrese el nombre del trabajador: ")
    edad = input("Ingrese la edad del trabajador: ")
    salario = input("Ingrese el salario del trabajador: ")
    nuevo_trabajador = {
        "ID": id_trabajador,
        "Nombre": nombre,
        "Edad": edad,
        "Salario": salario
    }
    lista.append(nuevo_trabajador)
    print("Trabajador agregado con éxito.")

def imprimir_lista(lista):
    print("Lista de trabajadores:")
    for trabajador in lista:
        print(trabajador)

def buscar_id(lista, id_buscar):
    for trabajador in lista:
        if trabajador["ID"] == id_buscar:
            print("Trabajador encontrado:")
            print(trabajador)
            return
    print("Trabajador no encontrado.")

def ordenar_lista(lista):
    lista.sort(key=lambda x: x["ID"])
    print("Lista ordenada por ID.")

def eliminar_id(lista, id_eliminar):
    for trabajador in lista:
        if trabajador["ID"] == id_eliminar:
            lista.remove(trabajador)
            print("Trabajador eliminado con éxito.")
            return
    print("Trabajador no encontrado.")

def borrar_toda_lista(lista):
    lista.clear()
    print("Lista de trabajadores borrada.")

def menu():
    lista_trabajadores = []
    while True:
        print("\nMENU\n")
        print("1.- Agregar (automático)")
        print("2.- Agregar (manual)")
        print("3.- Imprimir lista")
        print("4.- Buscar {ID}")
        print("5.- Ordenar")
        print("6.- Eliminar {ID}")
        print("7.- Borrar Toda la lista")
        print("0.- SALIR")
        opcion = input("Ingrese su opción: ")

        if opcion == "1":
            agregar_automatico(lista_trabajadores)
        elif opcion == "2":
            agregar_manual(lista_trabajadores)
        elif opcion == "3":
            imprimir_lista(lista_trabajadores)
        elif opcion == "4":
            id_buscar = input("Ingrese el ID del trabajador a buscar: ")
            buscar_id(lista_trabajadores, id_buscar)
        elif opcion == "5":
            ordenar_lista(lista_trabajadores)
        elif opcion == "6":
            id_eliminar = input("Ingrese el ID del trabajador a eliminar: ")
            eliminar_id(lista_trabajadores, id_eliminar)
        elif opcion == "7":
            borrar_toda_lista(lista_trabajadores)
        elif opcion == "0":
            print("Saliendo del programa...")
            break
        else:
            print("Opción no válida. Inténtelo de nuevo.")

menu()
```
